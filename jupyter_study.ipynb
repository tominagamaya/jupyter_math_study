```python
def main():
    num_str = input('好きな数字を入力してね：')
    num = int(num_str)
    furuits_series(num)
    
def furuits_series(num):
    fruits = ['りんご', 'ばなな', 'スイカ', 'いちご', 'ぶどう', 'キウイ']
    idx = int(num) % len(fruits)
    print('あなたが食べたいフルーツは')
    print(fruits[idx])
    
if __name__ == '__main__':
    main()
```

```python
from random import randint
num = randint(0, 7)
furuits_series(num)
```
